```python
def remove_markup(original_text):
  pattern = r'\'{2,5}'
  rm_text = re.sub(pattern, '', original_text)
  return rm_text
```

```python
import json
import re

filename = "/content/drive/MyDrive/Colab Notebooks/NLP100/chap03/data/enwiki-country.json"
with open(filename, mode='r') as f:
  for line in f:
    line = json.loads(line)
    if line["title"] == "United Kingdom":
      text = line["text"]
      break
# print(text)
pattern = r'^\{\{Infobox country.*?$(.*?)^\}\}'
template = re.findall(pattern, text, re.MULTILINE + re.DOTALL)

pattern = r'^\|(.+?)\s*=\s*(.+?)(?:(?=\n\|)|(?=\n$))'
result = dict(re.findall(pattern, template[0], re.MULTILINE + re.DOTALL))
result_rm = {k: remove_markup(v) for k, v in result.items()}
for k, v in result_rm.items():
    print(k + ': ' + v)
```
